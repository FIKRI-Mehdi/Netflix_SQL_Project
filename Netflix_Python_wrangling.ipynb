```python
import pandas as pd
import matplotlib as plt
import seaborn as sns
```

```python
# Importing the dataset
df = pd.read_csv(r"C:\Users\mehdi\OneDrive\Bureau\Data Analyst - courses\Project - Netflix\netflix_titles.csv")
df.head()
```

```python
#Initial data inspection
df.shape
df.columns
df.dtypes
df.isnull().sum()
```

```python
df['rating'].fillna('Unknown', inplace=True)
df['duration'].fillna('Unknown', inplace=True)
df.dropna(subset=['date_added'], inplace=True)
```

```python
df['director'].fillna('Unknown', inplace=True)
df['cast'].fillna('Unknown', inplace=True)
df['country'].fillna('Unknown', inplace=True)
```

```python
df.isnull().sum()
```

```python
# Data formatting 
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['title'] = df['title'].str.strip()
df = df[df['date_added'].notna()]
df.head()
```

```python
#Useful info extraction
df['date_added'] = pd.to_datetime(df['date_added'], format='%B %d, %Y')
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month

df['content_duration_int'] = df['duration'].str.extract('(\d+)').astype('float')
df['content_duration_type'] = df['duration'].str.extract('([a-zA-Z]+)')

#Converting the dates back to INT
df['year_added'] = df['year_added'].astype('Int64')
df['month_added'] = df['month_added'].astype('Int64')

df['content_duration_int'] = df['content_duration_int'].astype('Int64')
```

```python
df[['month_added','year_added','duration', 'content_duration_int', 'content_duration_type']].head()
```

```python
#Exporting the CSV
df.to_csv(r'C:\Users\mehdi\OneDrive\Bureau\Data Analyst - courses\Project - Netflix\netflix_cleaned.csv', index=False)
```
